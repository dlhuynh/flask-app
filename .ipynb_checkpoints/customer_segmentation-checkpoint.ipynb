```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# load data
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head(5)
data.tail(5)
```

```python
data.info()
```

```python
plt.figure(figsize=(15,5))
sns.distplot(data['Annual Income (k$)'])
plt.show()
```

```python
data.isnull().sum()
```

```python
# Convert gender to numerical
dummy=pd.get_dummies(data['Gender'], drop_first=True)
data1 = pd.concat([data[['Age','Annual Income (k$)','Spending Score (1-100)']],dummy],axis=1)
data1
```

```python
# find a cutoff value (Elbow method)
# as the best num of clusters from the range below
num_clusters_range=[2,3,4,5,6,7,8,9]
average_distance=[]
for n in num_clusters_range:
    km = KMeans(n_clusters = n,random_state=0)
    clust = km.fit(data1)
    average_distance.append(clust.inertia_)
plt.figure(figsize=(15,5))
plt.plot(num_clusters_range, average_distance)
plt.xlabel('Number of Clusters')
plt.ylabel('Mean Distance to Centroids')
plt.show()
```

```python
# Try Silhouette method
average_sh_scores=[]
for n in num_clusters_range:
    km = KMeans(n_clusters = n, random_state=0)
    clust_labels = km.fit_predict(data1)
    # average Silhouette across clusters
    avg_sh = silhouette_score(data1,clust_labels)
    average_sh_scores.append(avg_sh)
plt.figure(figsize=(15,5))
plt.plot(num_clusters_range, average_sh_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

### Results from the 2 different methods suggest the optimal number of clusters is 6

```python
# Looking into each cluster:
km = KMeans(n_clusters = 6, random_state=0)
clust_labels = km.fit_predict(data1)
centroids = km.cluster_centers_
```

```python
centroids
```

```python
data1['Cluster']=clust_labels
data1['Centroid_Age'] = data1['Cluster'].apply(lambda c: centroids[c][0])
data1['Centroid_Income'] = data1['Cluster'].apply(lambda c: centroids[c][1])
data1['Centroid_Spending'] = data1['Cluster'].apply(lambda c: centroids[c][2])
data1['Centroid_Gender'] = data1['Cluster'].apply(lambda c: centroids[c][3])
data1
```

```python
### Visualize clusters
plt.figure(figsize=(15,5))
colors = cm.nipy_spectral(clust_labels.astype(float)/6)
h=sns.scatterplot(data=data1, x="Spending Score (1-100)", y="Annual Income (k$)", hue='Cluster',c=colors)

plt.legend(loc='lower right', title='Cluster',bbox_to_anchor=(1.1, 0.2))
```

```python
plt.figure(figsize=(15,5))
colors = cm.nipy_spectral(clust_labels.astype(float)/6)
h=sns.scatterplot(data=data1, x="Spending Score (1-100)", y="Age", hue='Cluster',c=colors)

plt.legend(loc='lower right', title='Cluster',bbox_to_anchor=(1.1, 0.2))
```
